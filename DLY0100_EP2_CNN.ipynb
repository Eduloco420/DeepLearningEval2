## Por que las CNN son diferentes a las redes densas

Una red neuronal densa (MLP) trata cada pixel de una imagen como un dato independiente y lo conecta a todas las neuronas de la siguiente capa. Con imagenes de 224x224 pixels en 3 canales, eso son mas de 150,000 entradas y los parametros se vuelven inmanejables.

Una CNN aprovecha la **estructura espacial** de las imagenes. En vez de conectar todo con todo, aplica **filtros** (kernels) que se deslizan por la imagen buscando patrones locales: bordes, texturas, formas.

**La clave conceptual:** El mismo filtro que detecta un borde vertical en la esquina superior izquierda de la imagen sirve para detectarlo en cualquier otro lugar. A esto se llama **comparticion de parametros** y es lo que hace a las CNN eficientes.

## Que es una convolucion

Una convolucion es una operacion matematica donde un filtro (una matriz pequena, tipicamente 3x3 o 5x5) se desliza sobre la imagen multiplicando sus valores por los valores de la imagen en cada posicion y sumando el resultado.

El resultado se llama **mapa de caracteristicas** (feature map). Cada filtro aprende a detectar un tipo de patron especifico.

## La capa Conv2D

En Keras/TensorFlow, `Conv2D(filters, kernel_size, activation, padding)` es el bloque principal de una CNN.

**Parametros clave:**

- `filters`: Numero de filtros (cuantos patrones distintos busca la capa). Tipicamente 32, 64, 128, 256.
- `kernel_size`: Tamano del filtro. `(3,3)` es el mas comun.
- `activation`: Funcion de activacion. `'relu'` es el estandar.
- `padding`: `'same'` conserva las dimensiones espaciales. `'valid'` las reduce.

**Un ejemplo:** `Conv2D(32, (3,3), activation='relu', padding='same')` aplica 32 filtros de 3x3, produce 32 mapas de caracteristicas y conserva las dimensiones.

## La capa MaxPooling2D

El MaxPooling reduce las dimensiones espaciales tomando el valor maximo de cada region. Tiene dos efectos:

1. Reduce el numero de parametros en las siguientes capas.
2. Hace el modelo invariante a pequenos desplazamientos: si un borde se mueve un pixel, el maximo probablemente sigue siendo el mismo.

## Funciones de activacion

- **ReLU** (`max(0, x)`): La mas usada. Rapida, evita el problema del gradiente desvaneciente.
- **Sigmoid**: Salida entre 0 y 1. Util para clasificacion binaria.
- **Softmax**: Convierte los logits en probabilidades que suman 1. Usada en la capa de salida para clasificacion multiclase.

## Dropout y Batch Normalization

- **Dropout:** Desactiva aleatoriamente un porcentaje de neuronas durante el entrenamiento. Previene el sobreajuste (overfitting).
- **Batch Normalization:** Normaliza las activaciones de cada capa para estabilizar el entrenamiento y permitir tasas de aprendizaje mas altas.

````python
#@title Explorar parametros de Conv2D
import ipywidgets as widgets
from IPython.display import display, HTML
import numpy as np

BLUE, GOLD = '#1565C0', '#F9A825'

def calcular_dimensiones(entrada, filtros, kernel, padding, stride):
    if padding == 'same':
        h_out = int(np.ceil(entrada / stride))
        w_out = int(np.ceil(entrada / stride))
    else:  # valid
        h_out = int(np.floor((entrada - kernel) / stride)) + 1
        w_out = int(np.floor((entrada - kernel) / stride)) + 1
    params_filtro = kernel * kernel * filtros
    total_params = params_filtro + filtros  # biases
    total_activaciones = h_out * w_out * filtros
    html = f'''
    <div style="background:#f5f8ff;border:1px solid #1565C0;padding:16px 20px;border-radius:8px;font-family:monospace;">
      <p style="margin:2px 0;">Entrada: <strong>{entrada}x{entrada}</strong></p>
      <p style="margin:2px 0;">Filtros: <strong>{filtros}</strong> &nbsp;|&nbsp; Kernel: <strong>{kernel}x{kernel}</strong> &nbsp;|&nbsp; Padding: <strong>{padding}</strong> &nbsp;|&nbsp; Stride: <strong>{stride}</strong></p>
      <hr style="border-color:#1565C0;">
      <p style="margin:2px 0;">Salida: <strong style="color:#1565C0;">{h_out}x{w_out}x{filtros}</strong></p>
      <p style="margin:2px 0;">Parametros entrenables: <strong style="color:#F9A825;">{total_params:,}</strong></p>
      <p style="margin:2px 0;">Total de activaciones: <strong>{total_activaciones:,}</strong></p>
    </div>'''
    display(HTML(html))

w_entrada  = widgets.IntSlider(value=64, min=16, max=224, step=16, description='Entrada:', continuous_update=False)
w_filtros  = widgets.Dropdown(options=[8,16,32,64,128,256], value=32, description='Filtros:')
w_kernel   = widgets.Dropdown(options=[1,3,5,7], value=3, description='Kernel:')
w_padding  = widgets.Dropdown(options=['same','valid'], value='same', description='Padding:')
w_stride   = widgets.Dropdown(options=[1,2,3], value=1, description='Stride:')

out = widgets.interactive_output(
    calcular_dimensiones,
    {'entrada': w_entrada, 'filtros': w_filtros, 'kernel': w_kernel, 'padding': w_padding, 'stride': w_stride}
)

display(widgets.HTML('<p style="font-weight:bold;color:#1565C0;font-size:1.05em;">Calcula las dimensiones de salida de una capa Conv2D</p>'))
display(widgets.HBox([w_entrada, w_filtros, w_kernel]))
display(widgets.HBox([w_padding, w_stride]))
display(out)
````

````python
#@title Visualizar efecto del Dropout
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

BLUE, GOLD, BLACK = '#1565C0', '#F9A825', '#000000'
plt.rcParams['figure.facecolor'] = 'white'

np.random.seed(42)
activaciones_originales = np.random.randn(5, 8)

def mostrar_dropout(tasa):
    mask = np.random.rand(*activaciones_originales.shape) > tasa
    dropout_result = activaciones_originales * mask
    if tasa < 1.0:
        dropout_result = dropout_result / (1 - tasa)  # escala para mantener la esperanza

    fig, axes = plt.subplots(1, 2, figsize=(12, 4), facecolor='white')
    im1 = axes[0].imshow(activaciones_originales, cmap='Blues', aspect='auto', vmin=-2, vmax=2)
    axes[0].set_title('Activaciones originales', color=BLUE, fontweight='bold', fontsize=11)
    axes[0].set_xlabel('Neuronas', fontsize=10)
    axes[0].set_ylabel('Muestras del batch', fontsize=10)
    plt.colorbar(im1, ax=axes[0])

    im2 = axes[1].imshow(dropout_result, cmap='Blues', aspect='auto', vmin=-2, vmax=2)
    n_zeros = np.sum(mask == 0)
    n_total = mask.size
    axes[1].set_title(f'Despues de Dropout (tasa={tasa:.1f}) - {n_zeros}/{n_total} neuronas desactivadas', color=GOLD, fontweight='bold', fontsize=10)
    axes[1].set_xlabel('Neuronas', fontsize=10)
    plt.colorbar(im2, ax=axes[1])
    plt.tight_layout()
    plt.show()
    print(f'Tasa Dropout: {tasa:.1%} | Neuronas activas: {n_total - n_zeros}/{n_total} ({(1-tasa):.0%})')

tasa_widget = widgets.FloatSlider(
    value=0.3, min=0.0, max=0.9, step=0.1,
    description='Tasa Dropout:',
    style={'description_width': '120px'},
    layout=widgets.Layout(width='400px'),
    continuous_update=False
)
display(widgets.HTML('<p style="font-weight:bold;color:#1565C0;">Ajusta la tasa de Dropout y observa cuantas neuronas se desactivan:</p>'))
display(widgets.interactive(mostrar_dropout, tasa=tasa_widget))
````

# Preprocesamiento de imagenes CNN
## Por que necesitamos preprocesar las imagenes

Las imagenes en formato JPEG/PNG tienen pixeles con valores de 0 a 255. Antes de entrenar una CNN necesitamos:

1. **Normalizar los pixeles** a rango [0,1] o [-1,1] para estabilizar el gradiente.
2. **Redimensionar** todas las imagenes al mismo tamano (la CNN espera entradas de tamano fijo).
3. **Dividir en train/validation** para poder medir el rendimiento durante el entrenamiento.
4. Opcionalmente, aplicar **data augmentation** para aumentar la variedad del dataset.

## Data Augmentation

Data augmentation crea versiones modificadas de las imagenes originales durante el entrenamiento:
- **Volteo horizontal:** Homer de izquierda a derecha sigue siendo Homer.
- **Rotacion:** Un personaje ligeramente inclinado sigue siendo reconocible.
- **Zoom:** El personaje puede aparecer mas cerca o mas lejos.
- **Desplazamiento:** El personaje puede no estar perfectamente centrado.

Esto evita el sobreajuste y mejora la capacidad de generalizacion del modelo.

# Principios de diseno CNN
## Principios para disenar una CNN

Cuando disenamos una CNN, tomamos una serie de decisiones. Cada una tiene consecuencias sobre el rendimiento y el costo computacional.

### Profundidad de la red
Una red mas profunda puede aprender representaciones mas abstractas, pero es mas dificil de entrenar y puede sobreajustar si el dataset es pequeno.

### Numero de filtros por capa
Mas filtros significan mas capacidad de aprender patrones distintos. La practica comun es comenzar con pocos filtros (32 o 64) e ir duplicando en capas mas profundas.

### Tamano del kernel
El kernel 3x3 es el estandar actual. Captura patrones locales sin ser demasiado grande. Kernels 5x5 o 7x7 se usan menos frecuentemente.

### Regularizacion
Batch Normalization y Dropout evitan el sobreajuste. BatchNorm acelera el entrenamiento. Dropout fuerza a la red a no depender de ninguna neurona en particular.

### Stride vs MaxPooling
Ambos reducen la dimension espacial, pero MaxPooling es mas comun porque es interpretable: selecciona el valor mas activo de cada region.

| Decision | Opcion A | Opcion B | Cuando usar A |
|----|----|----|-----|
| Profundidad | Red profunda (5+ capas conv) | Red superficial (2-3 capas) | Dataset grande, problema complejo |
| Filtros | Muchos (128-512) | Pocos (32-64) | Mas clases, mas diversidad visual |
| Regularizacion | BatchNorm + Dropout | Solo Dropout | Siempre que sea posible |
| Reduccion espacial | MaxPooling | Stride=2 | Cuando quieres interpretabilidad |

````python
#@title Arquitectura A - CNN Baseline
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# ----------------------------------------------------------------
# ARQUITECTURA A: CNN BASELINE
# Red simple de referencia. Sirve como punto de partida.
# 3 bloques conv-pool, sin regularizacion adicional.
# ----------------------------------------------------------------
def crear_arquitectura_A(num_clases, img_size=(64, 64)):
    """
    Arquitectura A - Baseline CNN
    - 3 bloques Conv2D + MaxPooling
    - Sin BatchNormalization
    - Dropout leve (0.3)
    - Facil de entrenar, rapida, pero puede sobreajustar
    """
    model = keras.Sequential([
        # Bloque 1
        layers.Conv2D(32, (3,3), activation='relu', padding='same',
                      input_shape=(img_size[0], img_size[1], 3)),
        layers.MaxPooling2D(2, 2),

        # Bloque 2
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Bloque 3
        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Clasificador
        layers.Flatten(),
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_clases, activation='softmax')
    ], name='Arquitectura_A_Baseline')
    return model

# Mostrar resumen
modelo_A = crear_arquitectura_A(num_clases=42)
modelo_A.summary()

# Contar parametros
params_A = modelo_A.count_params()
print(f'\nParametros totales Arquitectura A: {params_A:,}')
````

````python
#@title Arquitectura B - CNN Mas Profunda
# ----------------------------------------------------------------
# ARQUITECTURA B: CNN MAS PROFUNDA
# Mas capas convolucionales, mayor capacidad de aprendizaje.
# Util cuando la red A tiene underfitting.
# ----------------------------------------------------------------
def crear_arquitectura_B(num_clases, img_size=(64, 64)):
    """
    Arquitectura B - CNN mas profunda
    - 4 bloques Conv2D (algunos con doble conv)
    - Sin BatchNormalization
    - Dropout moderado (0.4)
    - Mayor capacidad, mas lenta de entrenar
    """
    model = keras.Sequential([
        # Bloque 1
        layers.Conv2D(32, (3,3), activation='relu', padding='same',
                      input_shape=(img_size[0], img_size[1], 3)),
        layers.Conv2D(32, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Bloque 2
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Bloque 3
        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Bloque 4
        layers.Conv2D(256, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),

        # Clasificador
        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_clases, activation='softmax')
    ], name='Arquitectura_B_Profunda')
    return model

modelo_B = crear_arquitectura_B(num_clases=42)
modelo_B.summary()
params_B = modelo_B.count_params()
print(f'\nParametros totales Arquitectura B: {params_B:,}')
````

````python
#@title Arquitectura C - CNN con BatchNorm
# ----------------------------------------------------------------
# ARQUITECTURA C: CNN CON BATCH NORMALIZATION
# Incorpora BatchNorm despues de cada capa convolucional.
# Mas estable en el entrenamiento, converge mas rapido.
# ----------------------------------------------------------------
def crear_arquitectura_C(num_clases, img_size=(64, 64)):
    """
    Arquitectura C - CNN con BatchNormalization
    - 3 bloques Conv2D + BatchNorm + MaxPooling
    - BatchNorm normaliza las activaciones en cada mini-batch
    - Permite usar learning rates mas altos
    - Menor sobreajuste, entrenamiento mas estable
    """
    model = keras.Sequential([
        # Bloque 1: Conv + BN + Pool
        layers.Conv2D(32, (3,3), padding='same',
                      input_shape=(img_size[0], img_size[1], 3)),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2, 2),

        # Bloque 2: Conv + BN + Pool
        layers.Conv2D(64, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2, 2),

        # Bloque 3: Conv + BN + Pool
        layers.Conv2D(128, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.MaxPooling2D(2, 2),

        # Bloque 4: Conv + BN + Pool
        layers.Conv2D(256, (3,3), padding='same'),
        layers.BatchNormalization(),
        layers.Activation('relu'),
        layers.GlobalAveragePooling2D(),  # Alternativa a Flatten

        # Clasificador
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_clases, activation='softmax')
    ], name='Arquitectura_C_BatchNorm')
    return model

modelo_C = crear_arquitectura_C(num_clases=42)
modelo_C.summary()
params_C = modelo_C.count_params()
print(f'\nParametros totales Arquitectura C: {params_C:,}')
````

Entrenar un modelo CNN no es solo ejecutar un `.fit()`. Implica tomar decisiones sobre **como** se entrena: el optimizador, la tasa de aprendizaje, el numero de epocas, y como reaccionar cuando el rendimiento se estanca.

En esta seccion vas a entender que es cada hiperparametro y como afecta al entrenamiento.

**Hiperparametros que vamos a ver:**
- Optimizador (Adam, SGD, RMSprop)
- Tasa de aprendizaje (learning rate)
- Batch size
- Epocas y early stopping
- Callbacks de entrenamiento

# Que es cada hiperparametro
## Hiperparametros clave del entrenamiento

### Optimizador
El optimizador decide como actualizar los pesos de la red en cada iteracion. Los mas usados son:

| Optimizador | Descripcion | Cuando usarlo |
|----|----|----|
| Adam | Adaptativo, ajusta la tasa por parametro | Por defecto, excelente para CNN |
| SGD con momentum | Clasico, mas lento pero mas estable | Cuando quieres mas control |
| RMSprop | Similar a Adam, bueno para RNNs | Menos comun en CNN |

### Tasa de aprendizaje (learning rate)
Controla cuanto se actualiza cada peso en cada iteracion.
- **Muy alta**: el modelo no converge, los pesos oscilan
- **Muy baja**: el entrenamiento es extremadamente lento
- **Tipico para CNN**: entre 0.0001 y 0.001

### Batch size
Cuantas imagenes se procesan antes de actualizar los pesos.
- **Batch grande** (128-256): gradientes mas estables, requiere mas memoria
- **Batch pequeno** (16-32): mas ruido en el gradiente, puede ayudar a generalizar
- **Tipico**: 32

### Epocas
Cuantas veces recorre el modelo el dataset completo.
- No hay un numero magico. Se usa **Early Stopping** para parar cuando el rendimiento se estanca.

### Learning rate schedule
Reducir el learning rate cuando el entrenamiento se estanca ayuda al modelo a converger mejor.

````python
#@title Visualizar impacto del learning rate
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

BLUE = '#1565C0'
GOLD = '#F9A825'

def simular_entrenamiento(lr, ruido=0.15, epocas=30):
    """Simula curvas de entrenamiento segun el learning rate."""
    np.random.seed(42)
    x = np.arange(epocas)
    if lr > 0.1:
        # LR muy alta: oscilacion, no converge
        loss = 1.5 + 0.3 * np.sin(x * 0.8) + np.random.normal(0, 0.2, epocas)
        acc = 0.3 + 0.05 * np.sin(x * 0.5) + np.random.normal(0, 0.05, epocas)
    elif lr > 0.01:
        # LR alta: converge rapido pero puede oscilar
        loss = 1.8 * np.exp(-x * 0.12) + 0.2 + np.random.normal(0, ruido, epocas)
        acc = 0.85 * (1 - np.exp(-x * 0.15)) + 0.10 + np.random.normal(0, ruido/3, epocas)
    elif lr > 0.0005:
        # LR ideal
        loss = 1.5 * np.exp(-x * 0.10) + 0.15 + np.random.normal(0, ruido*0.5, epocas)
        acc = 0.88 * (1 - np.exp(-x * 0.12)) + 0.08 + np.random.normal(0, ruido/4, epocas)
    else:
        # LR muy baja: aprende muy lento
        loss = 1.5 * np.exp(-x * 0.03) + 0.5 + np.random.normal(0, ruido*0.3, epocas)
        acc = 0.6 * (1 - np.exp(-x * 0.04)) + 0.05 + np.random.normal(0, ruido/5, epocas)
    return x, np.clip(loss, 0, 2), np.clip(acc, 0, 1)

def graficar_lr(lr):
    x, loss, acc = simular_entrenamiento(lr)
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    fig.patch.set_facecolor('white')
    for ax in axes:
        ax.set_facecolor('white')
        for spine in ax.spines.values():
            spine.set_edgecolor('#cccccc')

    axes[0].plot(x, loss, color=BLUE, linewidth=2)
    axes[0].set_title(f'Loss (lr={lr})', fontweight='bold')
    axes[0].set_xlabel('Epoca'); axes[0].set_ylabel('Loss')

    axes[1].plot(x, acc, color=GOLD, linewidth=2)
    axes[1].set_title(f'Accuracy (lr={lr})', fontweight='bold')
    axes[1].set_xlabel('Epoca'); axes[1].set_ylabel('Accuracy')

    plt.tight_layout()
    plt.show()

lr_widget = widgets.SelectionSlider(
    options=[0.5, 0.1, 0.01, 0.001, 0.0001],
    value=0.001,
    description='Learning Rate:',
    style={'description_width': 'initial'},
    layout={'width': '500px'}
)

out = widgets.interactive_output(graficar_lr, {'lr': lr_widget})
display(lr_widget, out)
````

````python
#@title Configurar callbacks de entrenamiento
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

# ----------------------------------------------------------------
# EARLY STOPPING
# Detiene el entrenamiento cuando val_loss no mejora en 'patience' epocas
# Restaura los mejores pesos al terminar
# ----------------------------------------------------------------
early_stopping = EarlyStopping(
    monitor='val_loss',      # Que metrica observar
    patience=8,              # Cuantas epocas tolerar sin mejora
    restore_best_weights=True,  # Volver a los mejores pesos
    verbose=1
)

# ----------------------------------------------------------------
# REDUCE LR ON PLATEAU
# Reduce el learning rate cuando val_loss no mejora
# Ayuda al modelo a 'afinar' la solucion cuando ya esta cerca
# ----------------------------------------------------------------
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,       # Multiplicar LR por este factor
    patience=4,       # Epocas sin mejora antes de reducir
    min_lr=1e-7,      # LR minimo permitido
    verbose=1
)

# ----------------------------------------------------------------
# MODEL CHECKPOINT
# Guarda el modelo cuando mejora val_accuracy
# ----------------------------------------------------------------
checkpoint = ModelCheckpoint(
    filepath='mejor_modelo.keras',
    monitor='val_accuracy',
    save_best_only=True,
    verbose=1
)

CALLBACKS = [early_stopping, reduce_lr, checkpoint]

print('Callbacks configurados:')
print(f'  EarlyStopping: patience={early_stopping.patience}')
print(f'  ReduceLROnPlateau: factor={reduce_lr.factor}, patience={reduce_lr.patience}')
print(f'  ModelCheckpoint: guardando en mejor_modelo.keras')
````

````python
#@title Compilar y entrenar Arquitectura A
# ----------------------------------------------------------------
# IMPORTANTE: Esta celda requiere que los generadores de datos
# esten configurados (Seccion 3). Ejecutar en orden.
# ----------------------------------------------------------------

train_generator = train_gen
val_generator   = val_gen

LEARNING_RATE = 0.001
EPOCAS = 40

# Recompilar modelo A con el optimizador y la funcion de perdida
modelo_A.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy',
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall')]
)

print('Iniciando entrenamiento de Arquitectura A...')
print(f'Learning rate: {LEARNING_RATE}')
print(f'Epocas maximas: {EPOCAS}')
print(f'Early stopping patience: 8')
print('-' * 50)

historia_A = modelo_A.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCAS,
    callbacks=CALLBACKS,
    verbose=1
)

print('\nEntrenamiento finalizado.')
print(f'Epocas completadas: {len(historia_A.history["accuracy"])}')
print(f'Mejor val_accuracy: {max(historia_A.history["val_accuracy"]):.4f}')
````

# Explicacion de metricas
## Metricas de evaluacion para clasificacion

### Accuracy
Fraccion de predicciones correctas sobre el total.
```
Accuracy = Predicciones correctas / Total de predicciones
```
**Limitacion:** Engana cuando las clases estan desbalanceadas. Si el 90% de imagenes son de Homer, un modelo que siempre predice Homer tendra 90% de accuracy sin aprender nada.

### Precision
De todas las veces que el modelo dijo 'esta imagen es clase X', cuantas veces tenia razon.
```
Precision = Verdaderos Positivos / (Verdaderos Positivos + Falsos Positivos)
```

### Recall
De todas las imagenes que realmente son clase X, cuantas detecto el modelo.
```
Recall = Verdaderos Positivos / (Verdaderos Positivos + Falsos Negativos)
```

### F1-Score
Media armonica entre precision y recall. Util cuando quieres un unico numero que balancee ambas.
```
F1 = 2 * (Precision * Recall) / (Precision + Recall)
```

### Matriz de confusion
Tabla que muestra cuantas veces el modelo predijo cada clase vs la clase real. Las filas son la clase real, las columnas son la prediccion. Los valores en la diagonal son predicciones correctas.

````python
#@title Matriz de confusion visual
def graficar_confusion_matrix(y_real, y_pred, nombres_clases, top_n=15):
    """
    Grafica la matriz de confusion.
    Si hay muchas clases, muestra solo las top_n mas frecuentes.
    """
    from collections import Counter
    conteos = Counter(y_real)
    clases_top = [c for c, _ in conteos.most_common(top_n)]
    idx = np.isin(y_real, clases_top)

    y_real_sub = y_real[idx]
    y_pred_sub = y_pred[idx]

    nombres_top = [nombres_clases[i] for i in sorted(clases_top)]
    clases_sorted = sorted(clases_top)

    cm = confusion_matrix(y_real_sub, y_pred_sub, labels=clases_sorted)
    cm_norm = cm.astype(float) / cm.sum(axis=1, keepdims=True)

    fig, ax = plt.subplots(figsize=(12, 10))
    fig.patch.set_facecolor('white')
    ax.set_facecolor('white')

    sns.heatmap(
        cm_norm,
        annot=True,
        fmt='.2f',
        cmap='Blues',
        xticklabels=nombres_top,
        yticklabels=nombres_top,
        ax=ax,
        linewidths=0.5
    )
    ax.set_title(f'Matriz de confusion (top {top_n} clases, normalizada)',
                 fontsize=13, fontweight='bold', pad=15)
    ax.set_xlabel('Clase predicha', fontsize=11)
    ax.set_ylabel('Clase real', fontsize=11)
    plt.xticks(rotation=45, ha='right', fontsize=8)
    plt.yticks(rotation=0, fontsize=8)
    plt.tight_layout()
    plt.show()

graficar_confusion_matrix(y_real, y_pred, nombres_clases, top_n=15)

# Calcular accuracy global
acc_global = np.mean(y_real == y_pred)
print(f'\nAccuracy global en test: {acc_global:.4f} ({acc_global*100:.1f}%)')
````

### 1. Descripcion de la arquitectura

La arquitectura implementada corresponde a una red neuronal convolucional secuencial (CNN), compuesta por tres bloques convolucionales seguidos de un clasificador final. Cada bloque fue diseñado siguiendo la estructura Conv2D → BatchNormalization → ReLU → MaxPooling2D.

El primer bloque utiliza una capa Conv2D con 32 filtros de tamaño 3×3. Esta capa recibe imágenes de entrada de 64×64×3 y genera una salida de 64×64×32. Posteriormente, la operación MaxPooling reduce las dimensiones a 32×32×32. Este bloque cuenta con 896 parámetros entrenables y tiene como objetivo detectar características básicas, como bordes, colores predominantes y patrones simples presentes en los personajes.

El segundo bloque incorpora una capa Conv2D con 64 filtros de 3×3. La entrada corresponde a un tensor de 32×32×32 y la salida generada es de 32×32×64, la cual luego se reduce a 16×16×64 mediante MaxPooling. Este bloque posee 18.496 parámetros y permite identificar patrones más complejos, como formas faciales y contornos.

El tercer bloque utiliza 128 filtros de 3×3. Recibe una entrada de 16×16×64 y produce una salida de 16×16×128, que posteriormente se reduce a 8×8×128. Con 73.856 parámetros, este bloque se enfoca en capturar características de alto nivel, necesarias para diferenciar correctamente a los distintos personajes.

Finalmente, el clasificador utiliza una capa GlobalAveragePooling2D para transformar la salida de 8×8×128 en un vector de 128 elementos, evitando agregar parámetros adicionales. Luego se aplica una capa Dense de 128 neuronas junto con un Dropout de 0.4 para reducir el sobreajuste, terminando con una capa Dense de 42 neuronas y activación softmax para la clasificación multiclase.

En total, el modelo cuenta con 116.074 parámetros entrenables. Esta cantidad se mantuvo deliberadamente baja para disminuir el riesgo de overfitting considerando que el dataset utilizado contiene aproximadamente 7.000 imágenes.

### 2. Seleccion de herramientas

**Framework utilizado:** 

Para el desarrollo del modelo se utilizó TensorFlow 2.x junto con la API de Keras.

**Ventajas del framework elegido para este problema:**
- Ventaja 1: API de alto nivel intuitiva: Keras permite definir arquitecturas complejas con pocas líneas de código usando keras.Sequential y layers, reduciendo errores de implementación y facilitando la iteración rápida sobre distintas arquitecturas.

### 3. Decisiones de diseno

**Por que elegiste ese numero de capas convolucionales:**
Se decidió utilizar tres bloques convolucionales debido al tamaño de entrada de las imágenes (64×64). Cada operación de MaxPooling reduce las dimensiones espaciales a la mitad, siguiendo la secuencia 64→32→16→8. Incorporar un cuarto bloque habría reducido la resolución hasta 4×4, provocando una pérdida excesiva de información espacial relevante para diferenciar correctamente las 42 clases del problema.

**Por que elegiste ese numero de filtros:**
Respecto a la cantidad de filtros, se utilizó el patrón tradicional de duplicar filtros progresivamente entre bloques (32→64→128). Las primeras capas requieren menos filtros debido a que capturan patrones simples, mientras que las capas profundas necesitan mayor capacidad para representar combinaciones más complejas de características visuales.

**Que funcion de activacion usaste y por que:**
La función de activación utilizada fue ReLU (Rectified Linear Unit), aplicada después de cada capa BatchNormalization mediante layers.Activation('relu'). Esta función fue seleccionada debido a que es el estándar actual en redes convolucionales, principalmente porque evita el problema del desvanecimiento del gradiente presente en funciones como sigmoid o tanh, además de permitir entrenamientos más rápidos y estables. En la capa de salida se utilizó la función softmax, ya que permite convertir las 42 salidas del modelo en probabilidades cuya suma total es igual a 1, siendo adecuada para tareas de clasificación multiclase.

````python
#@title Actividad 2 - Entrenar segunda arquitectura
# ================================================================
# ACTIVIDAD 2: Entrena una segunda arquitectura para comparar
# ================================================================
# Instrucciones:
# 1. Define tu segunda arquitectura. Debe ser DIFERENTE a la primera.
#    Puede tener mas/menos capas, mas/menos filtros, BatchNorm, etc.
# 2. Entrena ambas con las mismas condiciones (mismo LR, mismos callbacks)
# 3. La comparacion debe ser justa (mismas condiciones de entrenamiento)

def crear_mi_cnn_v2(num_clases=NUM_CLASES, img_size=IMG_SIZE):
    """
    Mi segunda arquitectura CNN (Version 2).
    Diferencias respecto a V1:
    - 4 bloques convolucionales en lugar de 3
    - Sin BatchNormalization (para comparar su efecto)
    - Flatten en lugar de GlobalAveragePooling2D
    - Dos capas Dense en el clasificador (256 → 128)
    - Dropout espacial (SpatialDropout2D) dentro de los bloques Conv
    """
    model = keras.Sequential([

        # Bloque 1: más filtros desde el inicio para mayor capacidad
        # 64 filtros detectan más variantes de bordes y colores básicos
        layers.Conv2D(64, (3,3), activation='relu', padding='same',
                      input_shape=(img_size[0], img_size[1], 3)),
        layers.SpatialDropout2D(0.1),  # Dropout espacial: elimina mapas enteros aleatoriamente
        layers.MaxPooling2D(2, 2),     # 64x64 → 32x32

        # Bloque 2: misma cantidad de filtros (64), doble conv por bloque
        # Dos Conv seguidas sin pooling intermedio aprenden features más complejas
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.Conv2D(64, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),     # 32x32 → 16x16

        # Bloque 3: aumentamos a 128 filtros
        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.SpatialDropout2D(0.2),
        layers.MaxPooling2D(2, 2),     # 16x16 → 8x8

        # Bloque 4: último bloque antes del clasificador
        layers.Conv2D(128, (3,3), activation='relu', padding='same'),
        layers.MaxPooling2D(2, 2),     # 8x8 → 4x4

        # Flatten: convierte 4x4x128 = 2048 valores en vector plano
        # Contraste directo con GlobalAvgPool de V1
        layers.Flatten(),

        # Clasificador más profundo: dos capas Dense
        layers.Dense(256, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(num_clases, activation='softmax')

    ], name='Mi_CNN_V2')
    return model

mi_modelo_v2 = crear_mi_cnn_v2()
mi_modelo_v2.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy',
             keras.metrics.Precision(name='precision'),
             keras.metrics.Recall(name='recall')]
)

mi_historia_v2 = mi_modelo_v2.fit(
    train_generator,
    validation_data=val_generator,
    epochs=40,
    callbacks=CALLBACKS,
    verbose=1
)

graficar_historia(mi_historia_v2, titulo='Mi CNN V2 - Curvas de aprendizaje')
````

# Transfer Learning - Introduccion
## Transfer Learning: aprender de redes pre-entrenadas

Hasta ahora hemos entrenado nuestras redes desde cero. Existe una tecnica mas poderosa: **Transfer Learning**.

La idea es tomar una red que ya fue entrenada en millones de imagenes (como ImageNet) y reutilizar lo que aprendio.

### Por que funciona
Las primeras capas de cualquier CNN aprenden detectores de bordes, texturas y formas basicas que son utiles para cualquier tipo de imagen.
Solo las capas finales son especificas del problema original.

### Modelos pre-entrenados disponibles en Keras

| Modelo | Parametros | Top-1 Accuracy (ImageNet) |
|----|----|-----|
| MobileNetV2 | 3.4M | 71.3% |
| EfficientNetB0 | 5.3M | 77.1% |
| ResNet50 | 25.6M | 74.9% |
| InceptionV3 | 23.9M | 77.9% |
| VGG16 | 138M | 71.5% |

Para un dataset mediano como Simpsons, **MobileNetV2 o EfficientNetB0** son opciones ideales por su eficiencia.

### Estrategias
- **Feature extraction:** Congelar todas las capas del modelo base, solo entrenar el clasificador final.
- **Fine-tuning:** Descongelar las ultimas capas del modelo base y ajustar con un LR muy bajo.

# Diagnosticar sobreajuste
## Como diagnosticar overfitting en tus curvas

El sobreajuste (overfitting) ocurre cuando el modelo aprende de memoria los datos de entrenamiento en lugar de generalizar.

**Senales de sobreajuste en las curvas:**
- El accuracy de entrenamiento sube continuamente hasta valores muy altos (>95%).
- El accuracy de validacion se estabiliza o baja.
- La brecha entre train y val aumenta con las epocas.

**Senales de subajuste (underfitting):**
- Tanto el accuracy de train como de val son bajos.
- Las curvas se estabilizan en un valor bajo rapidamente.
- El modelo no puede aprender ni siquiera los datos de entrenamiento.

**Como corregir sobreajuste:**
1. Aumentar Dropout
2. Agregar BatchNormalization
3. Reducir la complejidad del modelo
4. Aplicar mas data augmentation
5. Aumentar el dataset

**Como corregir underfitting:**
1. Aumentar la profundidad del modelo
2. Agregar mas filtros
3. Reducir el Dropout
4. Entrenar por mas epocas
5. Usar un learning rate mas alto

# Interpretar la matriz de confusion
## Como leer una matriz de confusion

La matriz de confusion es una de las herramientas mas informativas para evaluar un clasificador.

**Estructura basica:**
- **Filas**: clase real
- **Columnas**: clase predicha
- **Diagonal**: predicciones correctas
- **Fuera de la diagonal**: errores

**Dos versiones de la matriz:**

| Version | Valores | Util para |
|----|----|-----|
| Absoluta | Numero de imagenes | Ver cuantos errores hay en total |
| Normalizada | Porcentaje (0 a 1) | Comparar clases con diferente cantidad de imagenes |

**Como interpretar:**

Si en la fila de 'homer_simpson' el valor mas alto esta en la columna 'homer_simpson', el modelo clasifica bien a Homer.

Si hay un valor alto en la fila de 'bart_simpson' en la columna 'homer_simpson', el modelo confunde a Bart con Homer con frecuencia.

**Por que ocurren confusiones:**
- Personajes con colores similares
- Pocas imagenes de una clase
- Imagenes de baja calidad o ambiguas
- Personajes que aparecen juntos en muchas imagenes del dataset

# Reflexion sobre las metricas
## Que metrica elegir y por que

Cuando presentas los resultados de tu modelo, es importante elegir la metrica correcta segun el contexto.

**Accuracy:** Buena cuando las clases estan balanceadas. En Simpsons hay ~42 clases y algunas tienen muchas mas imagenes que otras, por lo que accuracy sola puede ser enganosa.

**F1-score macro:** Trata a cada clase con igual importancia. Si una clase tiene 10 imagenes y otra 1000, ambas ponderan igual. Util cuando todas las clases importan por igual.

**F1-score weighted:** Pondera por la cantidad de imagenes de cada clase. Clases con mas imagenes tienen mas peso. Mas cercano al accuracy.

**Precision vs Recall:**
- Si el error mas grave es clasificar incorrectamente como X (falsos positivos): prioriza precision.
- Si el error mas grave es no detectar casos de X (falsos negativos): prioriza recall.

**Para el caso de Simpsons:**
La aplicacion no tiene consecuencias criticas de un tipo especifico de error, por lo que F1-score macro es una buena metrica de referencia.
Reporta accuracy + F1 para tener una vision completa.

````python
#@title Tabla interactiva de hiperparametros
import ipywidgets as widgets
from IPython.display import display, HTML

def calcular_params_conv(in_h, in_w, in_c, n_filtros, kernel, stride=1, padding='same'):
    """Calcula parametros de una capa Conv2D."""
    # Parametros: kernel_h * kernel_w * in_channels * n_filtros + n_filtros (bias)
    params = kernel * kernel * in_c * n_filtros + n_filtros
    if padding == 'same':
        out_h = int(np.ceil(in_h / stride))
        out_w = int(np.ceil(in_w / stride))
    else:  # valid
        out_h = int(np.floor((in_h - kernel) / stride)) + 1
        out_w = int(np.floor((in_w - kernel) / stride)) + 1
    return params, out_h, out_w

import numpy as np

# Calcular parametros para una arquitectura tipica de 64x64
print('Calculo de parametros por capa (imagen 64x64x3):')
print(f'{"Capa":<20} {"Entrada":<15} {"Salida":<15} {"Params":<12}')
print('-' * 62)

capas = [
    ('Conv2D_1 (32f,3x3)', 64, 64, 3, 32, 3),
    ('MaxPool_1 (2x2)', None, None, None, None, None),
    ('Conv2D_2 (64f,3x3)', 32, 32, 32, 64, 3),
    ('MaxPool_2 (2x2)', None, None, None, None, None),
    ('Conv2D_3 (128f,3x3)', 16, 16, 64, 128, 3),
    ('MaxPool_3 (2x2)', None, None, None, None, None),
    ('Dense_1 (256)',  None, None, None, None, None),
]

h, w, c = 64, 64, 3
for nombre, *args in capas:
    if 'Conv2D' in nombre:
        in_h, in_w, in_c, nf, k = args
        params, oh, ow = calcular_params_conv(in_h, in_w, in_c, nf, k)
        print(f'{nombre:<20} {str((in_h,in_w,in_c)):<15} {str((oh,ow,nf)):<15} {params:<12,}')
        h, w, c = oh, ow, nf
    elif 'MaxPool' in nombre:
        h, w = h//2, w//2
        print(f'{nombre:<20} -               {str((h,w,c)):<15} 0')
    elif 'Dense' in nombre:
        flat = h * w * c
        p = flat * 256 + 256
        print(f'{nombre:<20} {flat:<15,} 256             {p:<12,}')
````
